```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(x, y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```
